```python
# Performing standard imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
salary_data = pd.read_csv('Salary_Data.csv')
```

```python
salary_data.head()
```

```python
salary_data.info()
```

```python
sns.pointplot('YearsExperience','Salary',data=salary_data,)
```

```python
sns.scatterplot('YearsExperience','Salary',data=salary_data,)
```

```python
sns.lmplot('YearsExperience','Salary',data=salary_data)
```

```python
# Preprocessing data
X = salary_data[['YearsExperience']].values
y = salary_data[['Salary']].values
# Train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/3, random_state=42)
```

```python
# Linear model
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
# Fit linear model
lm.fit(X_train,y_train)
# Reshape your data either using array.reshape(-1, 1) if your data has a single feature or array.reshape(1, -1) 
# if it contains a single sample.
```

```python
# Prediction
y_predict = lm.predict(X_test)
```

```python
# Plot Traing set
plt.figure(figsize=(10,6))
plt.scatter(X_train,y_train,color='red')
plt.plot(y_predict)
plt.plot(y_test)
```
