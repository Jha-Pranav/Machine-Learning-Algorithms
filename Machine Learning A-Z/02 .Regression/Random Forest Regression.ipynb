```python
#Performing standard imports 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('Position_Salaries.csv')
```

```python
X = df[['Level']].values
y = df['Salary'].values
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfr = RandomForestRegressor(n_estimators=100)
```

```python
rfr.fit(X,y)
```

```python
sns.scatterplot(df['Level'],df['Salary'])
sns.lineplot(df['Level'],rfr.predict(X))
```

```python
# Projecting it into higher resolution
plt.figure(figsize=(10,6))
x_axis = np.arange(min(X),max(X),0.01)
x_axis = x_axis.reshape(len(x_axis),1)
plt.scatter(df['Level'],df['Salary'])
plt.plot(x_axis,rfr.predict(x_axis))
```
